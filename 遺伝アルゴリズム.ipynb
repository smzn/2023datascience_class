#遺伝アルゴリズム

遺伝的アルゴリズム（Genetic Algorithm, GA）は、進化生物学の自然選択と遺伝の原理を模倣した最適化手法です。GAは、特定の問題に対する最適解を探索するために、以下のプロセスを繰り返します。

### 基本概念

1. **個体（Individual）**: 解候補の一つを表します。各個体は遺伝子（Gene）の集合として表現されます。
2. **集団（Population）**: 個体の集合。GAは、複数の個体からなる集団を進化させることで最適解を探索します。
3. **適応度（Fitness）**: 各個体の良し悪しを評価するための値。問題に対する解の質を表します。
4. **選択（Selection）**: 次世代に進む個体を選ぶプロセス。適応度の高い個体が選ばれやすい。
5. **交叉（Crossover）**: 選択された個体ペアの遺伝子を組み合わせて新しい個体を生成するプロセス。
6. **突然変異（Mutation）**: 新しい個体の遺伝子をランダムに変更するプロセス。多様性を維持し、局所最適解から脱出するのに役立つ。

### GAの基本的な流れ

1. **初期集団の生成**: ランダムに初期個体を生成し、初期集団を形成します。
2. **適応度の評価**: 各個体の適応度を計算し、集団内の個体の評価を行います。
3. **選択**: 適応度に基づいて、次世代に進む個体を選びます。適応度の高い個体が選ばれやすくなります。
4. **交叉（Crossover）**: 選択された個体ペアの遺伝子を交叉させて新しい個体を生成します。
5. **突然変異（Mutation）**: 新しい個体に対して小さな変更を加えます。
6. **次世代集団の形成**: 新しい個体が次世代の集団を形成し、再び適応度の評価を行います。
7. **停止条件の確認**: 規定の世代数に達するか、適応度が十分に高い個体が見つかるまで、選択・交叉・突然変異のステップを繰り返します。

### GAの特徴

- **探索能力**: GAは大規模な探索空間を効率的に探索できるため、複雑な最適化問題に適しています。
- **多様性維持**: 交叉と突然変異によって個体の多様性を維持し、局所最適解に陥りにくくします。
- **適応性**: 問題の特性に応じて適応度関数や遺伝子構造を柔軟に設定できるため、多様な問題に適用可能です。

### 適用例

GAは多くの分野で応用されています。以下はその一部です。

- **組合せ最適化問題**: 巡回セールスマン問題（TSP）、ナップザック問題など。
- **機械学習**: ニューラルネットワークのハイパーパラメータの最適化。
- **エンジニアリング設計**: 構造設計の最適化、制御パラメータの調整。
- **経済学・金融**: ポートフォリオ最適化、株価予測モデルのパラメータ調整。

遺伝的アルゴリズムは、問題の特性や要求に応じて適応度関数や遺伝子操作をカスタマイズすることで、効果的に問題解決を支援する強力なツールです。

### ナップザック問題の定式化

ナップザック問題は、以下のように定式化されます：

- **与えられたもの**：
  - $n$ 個のアイテムがあり、各アイテム $i$ は重さ $w_i$ と価値 $v_i$ を持つ。
  - ナップザック（リュックサック）の容量（最大重量） $W$。

- **求めるもの**：
  - ナップザックに詰め込むアイテムの集合を選んで、その総価値を最大化する。
  - ただし、選んだアイテムの総重量がナップザックの容量 $W$ を超えないようにする。

この問題は次のように数学的に表現されます：

最大化する式：
$$ \text{maximize} \sum_{i=1}^{n} v_i x_i $$

制約条件：
$$ \sum_{i=1}^{n} w_i x_i \leq W $$

ここで、 $x_i$ は 0 または 1 の二値変数で、アイテム $i$ を選ぶ場合は 1、選ばない場合は 0 です。

### 遺伝的アルゴリズムの遺伝子構造

ナップザック問題における遺伝子構造は、各アイテムの選択状態を示すビット列（バイナリ列）です。ビット列の長さはアイテムの数 $n$ に等しく、各ビットはそのアイテムが選ばれているかどうかを示します。

例：
- アイテムが5つある場合、遺伝子は `[1, 0, 1, 0, 1]` のようになります。
- ここで、1はそのアイテムが選ばれていることを示し、0は選ばれていないことを示します。

### 目的関数

遺伝的アルゴリズムにおける目的関数は、個体（遺伝子）の適応度を評価するために使用されます。ナップザック問題の場合、目的関数は選択されたアイテムの総価値を計算し、そのアイテムの総重量がナップザックの容量 $W$ を超えないようにします。

具体的には、個体（遺伝子）を入力として受け取り、その適応度を返す関数です。ナップザック問題の目的関数は以下のように定義されます：

### 遺伝アルゴリズムのパラメタ

遺伝的アルゴリズム（GA）には、最適化プロセスを制御するためのいくつかの重要なパラメタがあります。これらのパラメタは、アルゴリズムの性能や結果に大きな影響を与えるため、適切に設定することが重要です。

1. **世代数（Number of Generations, NGEN）**:
    - アルゴリズムが停止するまでに繰り返す世代の数です。
    - 世代数が多いほど、より多くの進化を経てより良い解に収束する可能性がありますが、計算コストも増加します。

2. **個体数（Population Size, POP）**:
    - 初期集団に含まれる個体の数です。
    - 大きい集団サイズは多様性を保ちやすく、より良い解を見つける可能性が高まりますが、計算コストも増加します。

3. **交叉確率（Crossover Rate, CXPB）**:
    - 各世代で交叉を行う個体ペアの割合です。
    - 一般的には0.6〜0.9の範囲で設定されます。交叉確率が高いほど、遺伝子の組み合わせが多様化し、探索範囲が広がります。

4. **突然変異確率（Mutation Rate, MUTPB）**:
    - 各世代で突然変異を行う個体の割合です。
    - 一般的には0.01〜0.1の範囲で設定されます。突然変異確率が低すぎると局所最適解に陥りやすく、高すぎると収束が遅くなる可能性があります。

5. **選択方法（Selection Method）**:
    - 次世代に進む個体を選ぶ方法です。適応度の高い個体が選ばれやすくなるように設計されています。
    - 代表的な選択方法には以下のものがあります：
        - **ルーレット選択（Roulette Wheel Selection）**: 適応度に比例して個体が選ばれる確率を決定します。
        - **トーナメント選択（Tournament Selection）**: ランダムに選ばれた個体のグループから、最も適応度の高い個体を選びます。
        - **エリート選択（Elite Selection）**: 適応度の高い個体をそのまま次世代に残す方法です。

6. **交叉方法（Crossover Method）**:
    - 個体間で遺伝子を交叉させる方法です。交叉は親の遺伝子を組み合わせて新しい個体を生成します。
    - 代表的な交叉方法には以下のものがあります：
        - **一点交叉（One-Point Crossover）**: 遺伝子配列の一箇所で切断し、親の遺伝子を交叉させます。
        - **二点交叉（Two-Point Crossover）**: 遺伝子配列の二箇所で切断し、親の遺伝子を交叉させます。
        - **順序交叉（Order Crossover, OX）**: 特定の順序を保ちながら交叉を行います。巡回セールスマン問題（TSP）などで使用されます。

7. **突然変異方法（Mutation Method）**:
    - 突然変異を行う方法です。突然変異は遺伝子のランダムな位置を変更することで、多様性を維持し、局所最適解からの脱出を助けます。
    - 代表的な突然変異方法には以下のものがあります：
        - **ビット反転突然変異（Bit Flip Mutation）**: 二進数表現の遺伝子のビットをランダムに反転させます。
        - **シャッフル突然変異（Shuffle Mutation）**: 遺伝子の一部をランダムにシャッフルします。

これらのパラメタは、問題の特性や要求に応じて調整する必要があります。適切なパラメタ設定により、遺伝的アルゴリズムは効果的に最適解を探索し、優れた性能を発揮します。

### DEAPライブラリの説明

DEAP（Distributed Evolutionary Algorithms in Python）は、進化的アルゴリズムを実装するための強力かつ柔軟なPythonライブラリです。DEAPは、遺伝的アルゴリズム（GA）、進化戦略（ES）、遺伝的プログラミング（GP）など、さまざまな進化的アルゴリズムを簡単に構築、実装、カスタマイズするためのツールを提供します。

#### 特徴

- **柔軟性**: DEAPはモジュラー設計されており、さまざまな進化的アルゴリズムを簡単にカスタマイズできます。
- **シンプルさ**: シンプルかつ直感的なAPIを提供しており、使いやすさを重視しています。
- **並列処理**: ディストリビューションに基づく並列計算がサポートされており、計算を高速化できます。

#### 主なコンポーネント

DEAPにはいくつかの主要なコンポーネントがあり、これらを組み合わせて進化的アルゴリズムを構築します。

1. **Creator**:
    - 適応度（Fitness）と個体（Individual）を定義するために使用されます。
    - `creator.create`メソッドを使用して、新しいタイプを作成します。
    - 例:
      ```python
      from deap import creator
      creator.create("FitnessMax", base.Fitness, weights=(1.0,))
      creator.create("Individual", list, fitness=creator.FitnessMax)
      ```

2. **Toolbox**:
    - 遺伝的操作（選択、交叉、突然変異など）を登録するために使用されます。
    - `toolbox.register`メソッドを使用して、各操作を登録します。
    - 例:
      ```python
      from deap import base, tools
      toolbox = base.Toolbox()
      toolbox.register("mate", tools.cxTwoPoint)
      toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
      toolbox.register("select", tools.selTournament, tournsize=3)
      ```

3. **Algorithms**:
    - 進化的アルゴリズムの実行をサポートする高レベルのアルゴリズムが提供されています。
    - `eaSimple`、`eaMuPlusLambda`、`eaMuCommaLambda`などのアルゴリズムが用意されています。
    - 例:
      ```python
      from deap import algorithms
      algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=50, stats=stats, halloffame=hof, verbose=True)
      ```

4. **Stats**:
    - 進化の過程で集計された統計情報を収集するために使用されます。
    - 例:
      ```python
      stats = tools.Statistics(lambda ind: ind.fitness.values)
      stats.register("avg", np.mean)
      stats.register("std", np.std)
      stats.register("min", np.min)
      stats.register("max", np.max)
      ```

5. **HallOfFame**:
    - 最も優れた個体を保存するためのクラスです。
    - 例:
      ```python
      hof = tools.HallOfFame(1)
      ```



### GPAプロンプト : ナップザック問題(遺伝アルゴリズム利用)
* 遺伝アルゴリズムを使い、ナップザック問題での最適解を求める
* 遺伝アルゴリズムのライブラリはdeapを利用
* ナップザック問題を解くクラスとして、KnapSackDeapクラスを作成する
* ナップザック問題の情報は下記のリストを利用する
* name = ['チョコ', 'ポテトチップス', 'クッキー', 'ラムネ', 'ガム']
* weight = [130, 120, 80, 30, 20]
* value = [18, 15, 12, 4, 2]
* 制限容量は300とする
* 世代数（NGEN）は50, 個体数（POP）は80, 交叉確率（CXPB）は0.9, 突然変異確率（MUTPB）は0.1とする
* 目的関数の収束がわかるように、世代ごとの目的関数値（総価値）の推移を示すグラフを追加する（x軸を世代数、y軸を目的関数値とする）
* 最適解として、選ばれたアイテムの名前とその価値、総価値、および総重量を表示する

````python
import random
import numpy as np
import matplotlib.pyplot as plt
from deap import base, creator, tools, algorithms

class KnapSackDeap:
    def __init__(self, names, weights, values, capacity, pop_size=80, ngen=50, cxpb=0.9, mutpb=0.1):
        self.names = names
        self.weights = weights
        self.values = values
        self.capacity = capacity
        self.pop_size = pop_size
        self.ngen = ngen
        self.cxpb = cxpb
        self.mutpb = mutpb

        # DEAP setup
        creator.create("FitnessMax", base.Fitness, weights=(1.0,))
        creator.create("Individual", list, fitness=creator.FitnessMax)

        self.toolbox = base.Toolbox()
        self.toolbox.register("attr_bool", random.randint, 0, 1)
        self.toolbox.register("individual", tools.initRepeat, creator.Individual, self.toolbox.attr_bool, n=len(weights))
        self.toolbox.register("population", tools.initRepeat, list, self.toolbox.individual)

        self.toolbox.register("evaluate", self.evaluate)
        self.toolbox.register("mate", tools.cxUniform, indpb=0.5)
        self.toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
        self.toolbox.register("select", tools.selTournament, tournsize=3)

    def evaluate(self, individual):
        weight = sum(individual[i] * self.weights[i] for i in range(len(individual)))
        value = sum(individual[i] * self.values[i] for i in range(len(individual)))
        if weight > self.capacity:
            return 0,
        return value,

    def solve(self):
        pop = self.toolbox.population(n=self.pop_size)
        hof = tools.HallOfFame(1)
        stats = tools.Statistics(lambda ind: ind.fitness.values)
        stats.register("avg", np.mean)
        stats.register("std", np.std)
        stats.register("min", np.min)
        stats.register("max", np.max)

        log = algorithms.eaSimple(pop, self.toolbox, cxpb=self.cxpb, mutpb=self.mutpb, ngen=self.ngen,
                                  stats=stats, halloffame=hof, verbose=True)

        return pop, log, hof

    def print_solution(self, hof):
        best_ind = hof[0]
        total_weight = sum(best_ind[i] * self.weights[i] for i in range(len(best_ind)))
        total_value = sum(best_ind[i] * self.values[i] for i in range(len(best_ind)))
        selected_items = [self.names[i] for i in range(len(best_ind)) if best_ind[i] == 1]
        selected_values = [self.values[i] for i in range(len(best_ind)) if best_ind[i] == 1]

        print("Selected items:", selected_items)
        print("Selected values:", selected_values)
        print("Total value:", total_value)
        print("Total weight:", total_weight)

        return selected_items, selected_values, total_value, total_weight

    def plot_log(self, log):
        gen = log[1]
        max_fitness_values = [entry['max'] for entry in gen]

        plt.plot(max_fitness_values, label='Max Fitness')
        plt.xlabel('Generation')
        plt.ylabel('Max Fitness Value')
        plt.title('Fitness Evolution')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.show()
````

````python
# ナップザック問題の設定
names = ['チョコ', 'ポテトチップス', 'クッキー', 'ラムネ', 'ガム']
weights = [130, 120, 80, 30, 20]
values = [18, 15, 12, 4, 2]
capacity = 300

# KnapSackDeapインスタンスの作成と問題の解決
knapsack = KnapSackDeap(names, weights, values, capacity)
pop, log, hof = knapsack.solve()

# 解の表示
knapsack.print_solution(hof)

# グラフの表示
knapsack.plot_log(log)
````

### 巡回セールスマン問題（TSP）の説明と定式化

巡回セールスマン問題（Traveling Salesman Problem, TSP）は、組合せ最適化問題の一つであり、以下のように定義されます。

#### 説明

1. **問題の概要**：
    - 巡回セールスマン問題では、セールスマンが一つの都市から出発し、指定された複数の都市を一度ずつ訪問し、再び出発地点に戻る最短経路を求めることが目的です。

2. **応用例**：
    - 配送ルートの最適化
    - 工場の機械のメンテナンス順序の最適化
    - 観光地巡りの最適ルート決定



#### 定式化

巡回セールスマン問題は次のように数学的に定式化されます：

1. **与えられるもの**： $n$ 個の都市と、それぞれの都市間の距離（またはコスト） $d_{ij}$, ただし $i, j \in \{1, 2, ..., n\}$。

2. **求めるもの**：
    - 全ての都市を一度ずつ訪問し、出発点に戻る経路で、その総距離（コスト）を最小化する。

3. **数学的表現**：

    - **決定変数**：
        - $x_{ij}$：都市 $i$ から都市 $j$ へ移動する場合 $x_{ij} = 1$、それ以外は $x_{ij} = 0$。

    - **目的関数**：
        - 総移動距離を最小化する。
        $$ \text{minimize} \sum_{i=1}^{n} \sum_{j=1}^{n} d_{ij} x_{ij} $$

    - **制約条件**：
        - 各都市からはちょうど一度出発する。
        $$ \sum_{j=1, j \neq i}^{n} x_{ij} = 1 \quad \forall i \in \{1, 2, ..., n\} $$

        - 各都市へはちょうど一度到着する。
        $$ \sum_{i=1, i \neq j}^{n} x_{ij} = 1 \quad \forall j \in \{1, 2, ..., n\} $$

        - サブツアー（部分的な巡回路）が発生しないようにする（これを防ぐために、特別な制約が必要となることが多い）。

#### サブツアー除去制約の例

- **MTZ制約（Miller-Tucker-Zemlin制約）**：
    - 都市 $i$ を訪れたときの順序を $u_i$ として、次のような制約を追加する。
    $$ u_i - u_j + n x_{ij} \leq n - 1 \quad \forall i \neq j, i \neq 1, j \neq 1 $$
    - ただし、$u_i$ は2から$n$までの整数をとる（出発都市は1とする）。

巡回セールスマン問題はNP困難な問題であり、厳密な最適解を求めるには計算時間が指数関数的に増加します。そのため、現実的な応用では近似アルゴリズムやメタヒューリスティック手法（例：遺伝的アルゴリズム、シミュレーテッド・アニーリング、アント・コロニー・オプティマイゼーション）を用いて解決することが一般的です。

### GPTプロンプト
* 浦安駅を出発・帰着拠点として巡回セールス問題を解く
* 遺伝アルゴリズムで解く
* deapを利用する
* 遺伝子数=100, 世代数50として、世代ごとの最良遺伝子の目的関数値を折れ線グラフで示す
* 他のパラメタの値は一般的な値を利用
* 計算結果として、総移動距離を表示する(単位m)
* 移動経路を地図上に表示する(マーカ上に巡回経路番号を振る)
* 巡回経路経路を表示する
* 計算時間を計測し、表示する
* パラメタ設定は次のようにする
* NGEN = 50  # 世代数
* POP = 100  # 個体数
* CXPB = 0.9  # 交叉確率
* MUTPB = 0.1  # 突然変異確率

* Pythonコード

````python
import random
import numpy as np
import matplotlib.pyplot as plt
from deap import base, creator, tools, algorithms
import time

# 出発・帰着地点（浦安駅）のインデックス
start_index = df.index[df['場所名称'] == '浦安駅'][0]

# 距離行列を取得
distance_matrix = distance_df.values

# 遺伝的アルゴリズムの設定
NGEN = 50  # 世代数
POP = 100  # 個体数
CXPB = 0.9  # 交叉確率
MUTPB = 0.1  # 突然変異確率

creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

def evalTSP(individual):
    distance = 0
    for i in range(len(individual) - 1):
        distance += distance_matrix[individual[i]][individual[i + 1]]
    return distance,

def create_individual():
    cities = list(range(len(df)))
    cities.remove(start_index)
    random.shuffle(cities)
    return [start_index] + cities + [start_index]

def mutate_individual(individual, indpb):
    # スタートとゴールを除いた部分を突然変異させる
    part = individual[1:-1]
    if random.random() < indpb:
        tools.mutShuffleIndexes(part, indpb)
    individual[1:-1] = part
    return individual,

def cxOrderedFixed(ind1, ind2):
    # スタートとゴールを固定して一点交叉を行う
    part1 = ind1[1:-1]
    part2 = ind2[1:-1]
    tools.cxOrdered(part1, part2)
    ind1[1:-1] = part1
    ind2[1:-1] = part2
    return ind1, ind2

toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual, create_individual)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("mate", cxOrderedFixed)
toolbox.register("mutate", mutate_individual, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evalTSP)

# GAの実行
def main():
    random.seed(42)
    pop = toolbox.population(n=POP)
    hof = tools.HallOfFame(1)
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("min", np.min)
    stats.register("avg", np.mean)

    start_time = time.time()
    pop, logbook = algorithms.eaSimple(pop, toolbox, cxpb=CXPB, mutpb=MUTPB, ngen=NGEN, stats=stats, halloffame=hof, verbose=True)
    end_time = time.time()

    best_ind = hof[0]
    best_distance = best_ind.fitness.values[0]

    # 世代ごとの最良遺伝子の目的関数値をプロット
    gen = logbook.select("gen")
    fit_mins = logbook.select("min")
    plt.plot(gen, fit_mins, label="Minimum Fitness")
    plt.xlabel("Generation")
    plt.ylabel("Distance")
    plt.title("TSP Genetic Algorithm")
    plt.legend()
    plt.grid(True)
    plt.show()

    # 最適解の表示
    route = [df.iloc[i]['場所名称'] for i in best_ind]
    print(f"総移動距離: {best_distance} m")
    print(f"巡回経路: {' -> '.join(route)}")
    print(f"計算時間: {end_time - start_time} 秒")

    # 地図の中心を全拠点の重心に設定
    map_center = [df['緯度'].mean(), df['経度'].mean()]
    m = folium.Map(location=map_center, zoom_start=15)

    # データフレームの各行をマップにマーカーとして追加
    for idx, row in df.iterrows():
        if row['場所名称'] == '浦安駅':
            # 浦安駅のマーカー
            folium.Marker(
                location=[row['緯度'], row['経度']],
                popup=row['場所名称'],
                icon=folium.Icon(color='red', icon='info-sign')
            ).add_to(m)
        else:
            # その他拠点のマーカー
            folium.Marker(
                location=[row['緯度'], row['経度']],
                popup=row['場所名称']
            ).add_to(m)

    # 巡回経路を地図に追加
    for i in range(len(best_ind) - 1):
        folium.PolyLine(
            locations=[[df.iloc[best_ind[i]]['緯度'], df.iloc[best_ind[i]]['経度']],
                       [df.iloc[best_ind[i + 1]]['緯度'], df.iloc[best_ind[i + 1]]['経度']]],
            color='blue'
        ).add_to(m)
        folium.Marker(
            location=[df.iloc[best_ind[i]]['緯度'], df.iloc[best_ind[i]]['経度']],
            popup=f"{df.iloc[best_ind[i]]['場所名称']} (順序: {i + 1})",
            icon=folium.DivIcon(html=f'<div style="font-size: 12pt">{i + 1}</div>')
        ).add_to(m)

    # 最後のマーカー
    folium.Marker(
        location=[df.iloc[best_ind[-1]]['緯度'], df.iloc[best_ind[-1]]['経度']],
        popup=f"{df.iloc[best_ind[-1]]['場所名称']} (順序: {len(best_ind)})",
        icon=folium.DivIcon(html=f'<div style="font-size: 12pt">{len(best_ind)}</div>')
    ).add_to(m)

    # 地図を保存して表示
    map_path = "urayasu_bunkazai_tsp_ga_map.html"
    m.save(map_path)

    # 地図を表示
    return m

# 実行
if __name__ == "__main__":
    m = main()
````
